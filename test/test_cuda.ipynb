```python
import torch
import sys

print("=" * 50)
print("CUDA 验证工具")
print("=" * 50)

# 1. PyTorch 版本
print(f"\n1. PyTorch 版本: {torch.__version__}")

# 2. CUDA 是否可用
print(f"\n2. CUDA 可用: {torch.cuda.is_available()}")

if torch.cuda.is_available():
    # 3. CUDA 版本
    print(f"\n3. CUDA 版本: {torch.version.cuda}")
    
    # 4. GPU 信息
    print(f"\n4. GPU 信息:")
    print(f"   设备数量: {torch.cuda.device_count()}")
    for i in range(torch.cuda.device_count()):
        print(f"   设备 {i}: {torch.cuda.get_device_name(i)}")
        print(f"      显存: {torch.cuda.get_device_properties(i).total_memory / 1024**3:.1f} GB")
    
    # 5. 测试计算
    print(f"\n5. 测试 GPU 计算:")
    x = torch.randn(1000, 1000).cuda()
    y = torch.randn(1000, 1000).cuda()
    z = torch.matmul(x, y)
    print(f"   ✅ 矩阵乘法成功，结果形状: {z.shape}")
else:
    print(f"\n❌ CUDA 不可用，可能原因:")
    print("   1. 没有安装 NVIDIA 显卡驱动")
    print("   2. 安装了 CPU 版本的 PyTorch")
    print("   3. CUDA 版本不匹配")

print("\n" + "=" * 50)
```
